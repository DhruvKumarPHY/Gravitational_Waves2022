# 1}  Download the data for all three detectors involved with the GW170814 detection, generate a PSD for each, and make a plot

```python
%matplotlib inline
import gwpy
```

```python
from gwosc.datasets import event_gps
from gwpy.timeseries import TimeSeries

gps = event_gps('GW170814')
print("GW170814 GPS:", gps)

ldata = TimeSeries.fetch_open_data('L1', int(gps)-512, int(gps)+512, cache=True)
hdata = TimeSeries.fetch_open_data('H1', int(gps)-512, int(gps)+512, cache= True)
vdata = TimeSeries.fetch_open_data('V1', int(gps)-512, int(gps)+512, cache= True)
print("GW170814 data L1")
print(ldata)


print("GW170814 data H1")
print(hdata)


print("GW170814 data V1")
print(vdata)
```

```python
specgraml = ldata.spectrogram2(fftlength=4, overlap=2, window='hann') ** (1/2.)
plot = specgraml.plot()
specgramh = hdata.spectrogram2(fftlength=4, overlap=2, window='hann') ** (1/2.)
plot = specgramh.plot()
specgramv = vdata.spectrogram2(fftlength=4, overlap=2, window='hann') ** (1/2.)
plot = specgramv.plot()
```

```python
# Individual Events Tracing as per Detectors.
plot = specgraml.plot(vmin=1e-24, vmax=1e-20, norm='log')
al = plot.gca()
al.set_yscale('log')
al.set_ylim(10, 1400)
plot.add_colorbar(label=r'Strain noise [$1/\sqrt{\mathrm{Hz}}$]')


plot = specgramh.plot(vmin=1e-24, vmax=1e-20, norm='log')
ah = plot.gca()
ah.set_yscale('log')
ah.set_ylim(10, 1400)
plot.add_colorbar(label=r'Strain noise [$1/\sqrt{\mathrm{Hz}}$]')


plot = specgramv.plot(vmin=1e-24, vmax=1e-20, norm='log')
av = plot.gca()
av.set_yscale('log')
av.set_ylim(10, 1400)
plot.add_colorbar(label=r'Strain noise [$1/\sqrt{\mathrm{Hz}}$]')


# In this code we make use of specgraml.. function, hence, we get a classified data.
```

```python
# Combined Event Tracing From the Detectors
al = plot.gca()
al.set_yscale('log')
al.set_ylim(10, 1400)
al.colorbar(
    clim=(1e-24, 1e-20),
    norm="log",
    label=r"Strain noise [$1/\sqrt{\mathrm{Hz}}$]",
)
plot

ah = plot.gca()
ah.set_yscale('log')
ah.set_ylim(10, 1400)
ah.colorbar(
    clim=(1e-24, 1e-20),
    norm="log",
    label=r"Strain noise [$1/\sqrt{\mathrm{Hz}}$]",
)
plot

av = plot.gca()
av.set_yscale('log')
av.set_ylim(10, 1400)
av.colorbar(
    clim=(1e-24, 1e-20),
    norm="log",
    label=r"Strain noise [$1/\sqrt{\mathrm{Hz}}$]",
)
plot


# Here, we don't give any specgram argument. Hence, it gives a combined sum of the observation of 'GW170814'.
```
