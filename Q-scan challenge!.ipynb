# 1} Download the data for all detectors involved with the GW170814 detection. Which detectors registered this event?      Then, generate a Q-scan for the data in each detector. 


```python
%matplotlib inline
import gwpy
```

```python
from gwosc.datasets import event_gps
from gwpy.timeseries import TimeSeries

gps = event_gps('GW170814')
print("GW170814 GPS:", gps)

ldata = TimeSeries.fetch_open_data('L1', int(gps)-512, int(gps)+512, cache=True)
hdata = TimeSeries.fetch_open_data('H1', int(gps)-512, int(gps)+512, cache= True)
vdata = TimeSeries.fetch_open_data('V1', int(gps)-512, int(gps)+512, cache= True)
print("GW170814 data L1")
print(ldata)


print("GW170814 data H1")
print(hdata)


print("GW170814 data V1")
print(vdata)
```

```python
segmentl = (int(gps) - 30, int(gps) + 2)
ldata = TimeSeries.fetch_open_data('L1', *segmentl, verbose=True)
segmenth = (int(gps) - 30, int(gps) + 2)
hdata = TimeSeries.fetch_open_data('H1', *segmenth, verbose=True)
segmentv = (int(gps) - 30, int(gps) + 2)
vdata = TimeSeries.fetch_open_data('V1', *segmentv, verbose=True)
```

```python
lq = ldata.q_transform(frange=(20, 500), qrange=(4, 12),outseg=(gps-.1,gps+.05))
plot = lq.plot()
lx = plot.gca()
lx.set_epoch(gps)
lx.set_yscale('log')
lx.colorbar(label="Normalised energy")

hq = hdata.q_transform(frange=(20, 500), qrange=(4, 12),outseg=(gps-.1,gps+.05))
plot = hq.plot()
hx = plot.gca()
hx.set_epoch(gps)
hx.set_yscale('log')
hx.colorbar(label="Normalised energy")

vq = vdata.q_transform(frange=(20, 500), qrange=(4, 12),outseg=(gps-.1,gps+.05))
plot = vq.plot()
vx = plot.gca()
vx.set_epoch(gps)
vx.set_yscale('log')
vx.colorbar(label="Normalised energy")
```
